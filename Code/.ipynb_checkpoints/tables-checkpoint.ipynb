```python
import pandas as pd
```

```python
SLB_Data_all = pd.read_csv('../data/SLB/pmag_results.txt',sep='\t',skiprows=1)
```

```python
SLB_table_noparse = pd.DataFrame(SLB_Data_all.loc[SLB_Data_all['tilt_correction']==100.0], columns=['er_site_names',
                                                'average_alpha95', 
                                                'average_lat', 
                                                'average_lon', 
                                                'average_n', 
                                                'pole_comp_name','average_k', 
                                                'vgp_dm', 'vgp_dp', 
                                                'vgp_lat', 'vgp_lon'])
SLB_table_parse = pd.DataFrame(SLB_Data_all, columns=['er_site_names', 'average_dec',
                                                'average_inc','pole_comp_name', 'tilt_correction'])
#SLB_table_noparse = SLB_table_noparse.iloc[:-2]
SLB_table_parse = SLB_table_parse.iloc[:-2]
SLB_table_parse.set_index(['er_site_names', 'tilt_correction', 'pole_comp_name'], inplace=True)
SLB_table_parse = SLB_table_parse.unstack(level='tilt_correction')
SLB_table_parse.sort_index(level='pole_comp_name', ascending=False, inplace=True)
SLB_table_parse.sort_index(level='er_site_names', ascending=True, inplace=True, sort_remaining=False)
SLB_table_noparse.set_index(['er_site_names', 'pole_comp_name'], inplace=True)
# pd.merge(SLB_table.tilt_correction,SLB_table.loc[SLB_table['tilt_correction']==0.0], on='average_n')
SLB_table = SLB_table_parse.join(SLB_table_noparse)
# cols = SLB_table.columns.tolist()
# cols = cols[5:7] + [cols[7], cols[0], cols[2], cols[1], cols[3], cols[4], cols[8]] + cols[11:13] + cols[9:11]
# SLB_table = SLB_table[cols]
SLB_table = SLB_table[[5, 6, 7, 0, 2, 1, 3, 4, 8, 11, 12, 9, 10]]
```

```python
SLB_table
```

```python
print SLB_table.to_latex(column_format='c'*len(SLB_table.columns), bold_rows=True)
```

```python
All_Michi = pd.read_csv('../data/Michipicoten/pmag_results.txt',sep='\t',skiprows=1)
All_Michi = All_Michi.loc[All_Michi['pole_comp_name'] == 'HT']
All_Michi.reset_index(inplace=True,drop=True)
All_Michi
```

```python
Michi_table_noparse = pd.DataFrame(All_Michi.loc[All_Michi['tilt_correction']==100.0], columns=['er_site_names',
                                                'average_alpha95', 
                                                'average_lat', 
                                                'average_lon', 
                                                'average_n', 'average_k', 
                                                'vgp_dm', 'vgp_dp', 
                                                'vgp_lat', 'vgp_lon'])
Michi_table_parse = pd.DataFrame(All_Michi, columns=['er_site_names', 'average_dec',
                                                'average_inc', 'tilt_correction'])
#SLB_table_noparse = SLB_table_noparse.iloc[:-2]
#Michi_table_parse = Michi_table_parse.iloc[:-2]
Michi_table_parse.set_index(['er_site_names', 'tilt_correction'], inplace=True)
Michi_table_parse = Michi_table_parse.unstack(level='tilt_correction')
Michi_table_noparse.set_index(['er_site_names'], inplace=True)
# pd.merge(SLB_table.tilt_correction,SLB_table.loc[SLB_table['tilt_correction']==0.0], on='average_n')
Michi_table = Michi_table_parse.join(Michi_table_noparse)
rows = Michi_table.index.tolist()
rows = [rows[0], rows[1], rows[2], rows[13]] + rows[16:] + rows[3:13] +rows[14:16]
Michi_table = Michi_table.loc[rows]
# cols = Michi_table.columns.tolist()
# cols = [cols[0], cols[2], cols[1]] + cols[3:]
```

```python
Michi_table = Michi_table[[5, 6, 7, 0, 2, 1, 3, 4, 8, 11, 12, 9, 10]]
Michi_table
```

```python
print Michi_table.to_latex(column_format='c'*len(Michi_table.columns), bold_rows=True)
```
